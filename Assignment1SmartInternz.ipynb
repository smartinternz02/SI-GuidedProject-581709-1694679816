```python
import pandas as pd
import numpy as np
import random
```

```python
# Create a dictionary with random data for the 5 features
data = {
    'OrderID': [random.randint(1000, 1999) for _ in range(10)],
    'ShipMode': [random.choice(['FirstClass', 'SecondClass','StandardClass']) for _ in range(10)],
    'Segment': [random.choice(['Consumer', 'Corporate', 'HomeOffice']) for _ in range(10)],
    'Quantity': [random.randint(10, 200) for _ in range(10)],
    'Profit': [random.randint(500, 2000) for _ in range(10)]
}

# Create the DataFrame
df = pd.DataFrame(data)
```
