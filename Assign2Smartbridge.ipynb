```python
sns.pairplot(d1)
```

```python
corrmat = df.corr()   # corr() function calculates the Pearson correlation coefficients between pairs of features
k = 14  # sets the number of features to include in the heatmap.
cols = corrmat.nlargest(k, 'Price')['Price'].index   # selects the top k features with the highest correlation values with the target variable ('status')
cm = np.corrcoef(df[cols].values.T)  # np.corrcoef() calculates the correlation matrix for the transposed values of the selected features.
plt.figure(figsize=(14,12))    # specified size to display the heatmap
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values, cmap='YlGn', linecolor='white')
```
